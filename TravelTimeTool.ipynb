## TravelTimeTool

User can provide the input parameter below to run the tool. The input format is summarized below:
- `year_ykr`: List. Each item should have the year and YKR ID seperated by "_" \
   e.g ["2018_5785642","2015_5785642"]

- `vector_type`: String. Either "shp" for Shapefile or "gpkg" for Geopackage

- `vector_output_folder`: String. The output folder name

- `map_type`: String. Either "static" for static map or "interactive" for interactive map

- `map_output_folder`: String. The output folder name

- `transport_mode`: String. the transport mode of interest

- `mode_cmp`: List or None. Each item should have the transport mode of interest \
e.g ["walk_t","pt_m_t"]

- `grid_cmp`: List or None. Each item should have the year, ykr_id and transport mode of interest.\
The travel time data file must also be specified in `year_ykr`.\
e.g ["2018_5785642_walk_t","2015_5785642_pt_m_t"]

- `interactive_map_display`:\
`map_type` = `static`: None \
`map_type` = `interactive`: "Yes" or "No"

```python
from function import access_viz

access_viz (
    year_ykr=["2018_6016699","2015_6016699","2013_5785643"],
    vector_type="gpkg",
    vector_output_folder=DATA_DIRECTORY / "output",
    map_type="static",
    map_output_folder=DATA_DIRECTORY / "output",
    transport_mode="walk_t",
    mode_cmp=["walk_d","pt_m_d"],
    grid_cmp=["2018_6016699_car_m_t","2015_6016699_car_m_t"],
    interactive_map_display=None
)
```
